#### [데이터 필터링]
- Series/DataFrame에서 조건에 맞는 데이터 추출
- 방법
    *Boolean Indexing Series
    *isin() / query()
- 필터링 연산자
    * 비교연산자: >, >=, <, <=, ==, !=
    * 논리연산자: &(and), |(or), ~(not)

[1] 모듈 로딩 및 데이터 준비 <hr>

```python
import pandas as pd
```

```python
DATA_FILE = r'..\Data\titanic.csv'

dataDF = pd.read_csv(DATA_FILE)

## 20개 행만 선택
#dataDF = dataDF.iloc[:20, :7]
dataDF = dataDF.loc[:19, 'survived':'fare']

display(dataDF, dataDF.columns, dataDF.shape)
```

[2] 불린 인덱싱(Boolean Indexing) <hr>

```python
## [조건] 생존자 데이터만 추출
## - DF[컬럼명]: 컬럼 선택
retSR = dataDF['survived'] == 1
print(retSR)

## - 전체 DF에서 A컬럼 데이터가 20이상인 데이터만 선택
dataDF1 = dataDF[retSR]
dataDF2 = dataDF[dataDF['survived'] == 1]

## 결과 확인: 생존자의 나이와 성별만 확인
print(dataDF1[['age', 'sex']], dataDF2[['age', 'sex']], sep="\n\n")
```

```python
## [조건] 생존자 중에서 30대 남자인 데이터만 선택
## 생존자 == 1
## 나이: 30 <= ~ < 40
## 성별 == 'male'
mask = (dataDF.survived == 1) & ((dataDF.age >= 20) & (dataDF.age < 30)) & (dataDF.sex == 'female')

dataDF[mask][['survived', 'sex', 'age']]
```
