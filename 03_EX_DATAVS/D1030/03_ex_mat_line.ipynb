### [선 그래프]
- 시간의 흐름에 따른 추세/변화/트렌드 표현

[1] 모듈 로딩 및 데이터 준비 <hr>

```python
## 모듈 로딩
import pandas as pd             # 데이터 분석 관련 모듈
import matplotlib.pyplot        # 그래프 관련 모듈
import koreanize_matplotlib     # 한글 폰트 관련 모듈
```

```python
## 데이터 준비
data = {'A':[10, 20, 30, 40, 50], 'B':[15, 25, 35, 45, 55]}

dataDF = pd.DataFrame(data)
display(dataDF)
```

[2] A에 따른 B의 변화 시각화 <hr>

```python
import matplotlib.pyplot as plt
## -------------------------------------------------------------------
## 선그래프 함수: plt.plot(x, y, fnt)
## -------------------------------------------------------------------
plt.plot(dataDF['A'], dataDF['B'], 'r-')

## 그래프 기본 설정
plt.xlabel("A")
plt.ylabel("B")
plt.title("[A에 따른 B의 변화]")

plt.show()
```

```python
import matplotlib.pyplot as plt
## -------------------------------------------------------------------
## 선그래프 함수: plt.plot(y, fmt)
## -------------------------------------------------------------------
plt.plot(dataDF['A'], 'r-')             # red     실선
plt.plot(dataDF['A'] + 10, 'b--')       # blue    점선
plt.plot(dataDF['A'] - 10, 'g^')        # green   삼각형 마크 + 선 X
plt.plot(dataDF['A'] - 5, 'm^-')        # magenta 삼각형 마크 + 실선

## 그래프 기본 설정
plt.ylabel("A")
plt.title("[A 변화]")

## 그래프 화면 출력
plt.show()
```

```python
import matplotlib.pyplot as plt
## -------------------------------------------------------------------
## 선그래프 함수: plt.plot(y, fmt, label)
## -------------------------------------------------------------------
plt.plot(dataDF['A'], 'r-', label='A')             # red     실선
plt.plot(dataDF['A'] + 10, 'b--', label='A + 10')       # blue    점선
plt.plot(dataDF['A'] - 10, 'g^', label='A - 10')        # green   삼각형 마크 + 선 X
plt.plot(dataDF['A'] - 5, 'm^-', label='A - 05')        # magenta 삼각형 마크 + 실선

## 그래프 기본 설정
plt.ylabel("A")
plt.title("[A 변화]")
plt.legend()                                            # 범례 출력

## 그래프 화면 출력
plt.show()
```
