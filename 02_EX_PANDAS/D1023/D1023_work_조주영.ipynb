```python
## 모듈 로딩
import pandas as pd
```

```python
data = [["2차전지(생산)", "SK이노베이션", 10.19, 1.29],
        ["해운", "팬오션", 21.23, 0.95],
        ["시스템반도체", "티엘아이", 35.97, 1.12],
        ["해운", "HMM", 21.52, 3.20],
        ["시스템반도체", "아이에이", 37.32, 3.55],
        ["2차전지(생산)", "LG화학", 83.06, 3.75]
]
columns = ["테마", "종목명", "PER", "PBR"]

```

```python
## (1-1) 종목명, PER, PBR 컬럼만 선택 후 DataFrame으로 저장하는 코드 작성하세요.
dataDF1 = pd.DataFrame(data, columns=columns)

display(dataDF1)
```

```python
##  (1-2)  PER이 20미만, 50이상인 데이터만 저장하는 코드 작성하세요.
retSR1 = (dataDF1['PER'] < 20) | (dataDF1['PER'] >= 50)

dataDF1[retSR1]
```

```python
## (1-3) 종목명을 행인덱스롤 설정하는 코드 작성하세요.
dataDF1_3 = dataDF1.set_index(['종목명'])
display(dataDF1_3)
```

```python

##  (1-4) PBR 컬럼의 데이터를 기준으로 내림차순 정렬해 주세요.
dataDF1_4 = dataDF1.sort_values(by='PBR', ascending=False)

display(dataDF1_4)
```

```python
##  (1-5) PER, PBR 순서로 오름차순 정렬해 주세요.
dataDF1_5 = dataDF1.sort_values(by=['PER', 'PBR'])

display(dataDF1_5)
```

```python
FILE5 = '../Data/titanic.csv'
df5 = pd.read_csv(FILE5)

display(df5)
```

```python
##  (5-1) 단순 조건에 맞는 데이터만 선택 
## 성별이 여자인 승객 행만 필터링.
cond1 = df5['sex'] == 'female'

display(df5[cond1])

## pclass가 1등석인 승객만.
cond2 = df5['pclass'] == 1

display(df5[cond2])
## age가 10세 미만인 어린이만.
cond3 = df5['age'] < 10

display(df5[cond3])
## fare가 100이상인 고가 요금 승객만.
cond4 = df5['fare'] >= 100

display(df5[cond4])
```

```python
##  (5-2) 아래 조합 조건에 맞는 데이터만 선택
## 1 또는 2등석인 생존자 데이터만
cond5 = (((df5['pclass'] == 1) | (df5['pclass'] == 2)) & (df5['survived'] == 1)) 

display(df5[cond5])
## 가족 동승 3명 이상) 승객 데이터만
cond6 = (df5['sibsp'] + df5['parch']) >= 3

display(df5[cond6])
## 혼자 탑승한 성인 남자 데이터만
cond7 = (df5['alone'] == True) & (df5['adult_male'] == True)

display(df5[cond7])
```
