### 【 행/열 선택 실습】


[1] 모듈 로딩 및 데이터 준비 <hr>

```python
import pandas as pd
```

```python
## 데이터 준비
DATA_FILE1 = '../Data/data01.csv'
DATA_FILE2 = '../Data/data02.csv'
```

```python
# 데이터 로딩 및 DataFrame 변환 저장
dataDF1 = pd.read_csv(DATA_FILE1, header=None)

#-> index_col 매개변수: 파일의 특정 컬럼을 행 인덱스로 지정
dataDF2 = pd.read_csv(DATA_FILE2, index_col=0)
```

```python
display(dataDF1, dataDF2)
```

[dataDF1 실습 - 행과 열 선택] <hr>

```python
## -----------------------------------------------------------
## 행 선택
## -----------------------------------------------------------
## (1) 마지막 행 선택
last_row = dataDF1.iloc[-1]    # 또는 .loc[7] / 여기서 7은 인덱싱이 아닌 라벨을 의미

display(last_row)
## (2) 2번, 3번, 4번 행 선택
rows = dataDF1.iloc[[2, 3, 4]]    # 또는 .loc[[2, 3, 4]]

display(rows)
## (3) 홀수번째 행 선택
odd_rows = dataDF1.iloc[1:8:2]    # 또는 .loc[1:7:2]

display(odd_rows)
```

```python
## -----------------------------------------------------------
## 열 선택
## -----------------------------------------------------------
## (1) 첫번째 컬럼 선택
# first_col = dataDF1.0    # index가 str일 때만    
first_col = dataDF1[0]

display(first_col)

## (2) 2번, 3번 컬럼 선택 => List 방식
two_cols = dataDF1[[1, 2]]

display(two_cols)

## (3) 2번, 3번 컬럼 선택 => slicing 방식
two_cols_2 = dataDF1.iloc[:, 1:3]    # 또는 .loc[:, 1:2]

display(two_cols_2)
```

[dataDF2 실습 - 행과 열 선택] <hr>

```python
## -----------------------------------------------------------
## 행 선택
## -----------------------------------------------------------
## (1) 마지막 행 선택
last_row = dataDF2.iloc[-1]

display(last_row)
## (2) 2번, 3번, 4번째 행 선택 => 4, 5, 6
rows = dataDF2.iloc[[1, 2, 3]]

display(rows)

## (3) 홀수번째 행 선택
odd_rows = dataDF2.iloc[1::2]

display(odd_rows)
```

```python
## -----------------------------------------------------------
## 열 선택
## -----------------------------------------------------------
## (1) 첫번째 컬럼 선택
first_col = dataDF2['v1']

display(first_col)

## (2) 2번, 3번 컬럼 선택 => List 방식
two_cols = dataDF2[['v2', 'v3']]

display(two_cols)

## (3) 2번, 3번째 컬럼 선택 => slicing 방식
two_cols_2 = dataDF2.iloc[:, 1:3]

display(two_cols_2)
```
