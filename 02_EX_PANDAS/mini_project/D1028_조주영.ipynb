### [보조금과 전기차 보급의 상관관계]
- 전기차 등록현황 데이터를 기본 데이터로 사용
- 보조금 데이터를 통해 2010년대 후반의 전기차 등록의 급증 원인을 분석한다

```python
## 모듈 로딩
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
## 데이터 준비
FILE_NAME = '../project/DATA/전기차_등록현황.csv'

ev_carDF = pd.read_csv(FILE_NAME)
display(ev_carDF.head())
```

```python
## 데이터 정보
ev_carDF.info()

## 기본 통계
ev_carDF.describe()
```

```python
## 인덱스를 년월로, 2017년에서 2018년 데이터만을 추출
ev_carDF1 = ev_carDF.set_index('년월')
ev_carDF1 = ev_carDF1.loc['2018-12':'2017-01', '서울':'제주']

## 인덱스가 '2017-12' 형식의 문자열일 경우, 문자열 슬라이스로 연도 추출
ev_carDF1['연도'] = ev_carDF1.index.str[:4]

## 연도 그룹 내에서 평균 내기
ev_year_mean = ev_carDF1.groupby('연도').mean().round(1)

## 결과 확인
display(ev_year_mean)
```

```python
## 행-열 뒤집어서 증가율 컬럼 추가
ev_year_mean = ev_year_mean.T

## 시도로 인덱스 주기
ev_year_mean = ev_year_mean.rename_axis(['시도'], axis=0)

## 증가대수 추가
ev_year_mean['증가대수'] = (ev_year_mean['2018'] - ev_year_mean['2017']).round(1)
display(ev_year_mean)
```

[2] 전기차 보조금의 민간공고대수 데이터를 활용
- 민간공고대수와 등록 현황 증가의 상관관계를 확인한다.

```python
## (1) 전기차 민간공고대수 데이터 준비
SUB_FILE = '../project/DATA/보조금.csv'

ev_subDF = pd.read_csv(SUB_FILE, encoding='CP949')      ## CP949로 인코딩
## (2) 필요한 컬럼만 선택
ev_subDF = ev_subDF[['시도', '지역구분', '민간공고대수']]

## (3) 시도별 합계 계산
ev_sumDF = ev_subDF.groupby('시도', as_index=True)['민간공고대수'].sum().round(1)       ## 도는 시, 군으로 나뉘어져 있어서 sum

## (4) 등록 현황 데이터에 맞게 정렬
order = ['서울','부산','대구','인천','광주','대전','울산','세종',
         '경기','강원','충북','충남','전북','전남','경북','경남','제주']

ev_sumDF = ev_sumDF.reindex(order)

display(ev_sumDF)
```

```python
## 등록 현황과 보조금 공고대수 join
merged_df = ev_year_mean.join(ev_sumDF, how='left')     ## 왼쪽으로 merge

display(merged_df)
```

```python
# (1) 공고대수 대비 증가 비율 계산
merged_df['공고대수대비증가율(%)'] = ((merged_df['증가대수'] / merged_df['민간공고대수']) * 100).round(1)
mean_ratio = merged_df['공고대수대비증가율(%)'].mean().round(1)


# (2) 가중평균 계산 (증가대수를 가중치로)
weighted_mean = (
    (merged_df['공고대수대비증가율(%)'] * merged_df['증가대수']).sum()
    / merged_df['증가대수'].sum()
).round(2)

# (3) 상관계수 계산
corr = merged_df[['민간공고대수', '증가대수']].corr().iloc[0,1]
display(merged_df)

print(f"시도별 공고대수 대비 증가율 단순 평균: {mean_ratio}%")
print(f"시도별 공고대수 대비 증가율 가중 평균: {weighted_mean}%")
print(f"공고대수와 증가대수 간 상관계수     : {corr:.2f}")
```
